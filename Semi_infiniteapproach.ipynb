```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

patch_amount = 21 #mg
nicotine_molar_mass = 162.23 #g/mol
thickness = 0.037 #cm
area = 3 * 4.2 # cm^2
pi = 3.1415926535

patch_volume = area * thickness #cm^3
c0 = ((patch_amount * 10**-3) / nicotine_molar_mass) / patch_volume
c1 = 0

# Dij_skin = 2.56*10**-8 #cm^2/s
Dij_memb = 4.39*10**-10 #cm^2/s

t_used = 24 * 60 * 60 #s
print(t_used)

#Semi-infinite approach validation
t_needed = thickness**2 / (26 * Dij_memb)
print(t_needed)

#Flux w/ semi-inf approach

def flux_calc(Dij_memb, t, c1, c0):
  flux = -1 * (Dij_memb/(pi*t))**(1/2) * (c1 - c0)
  return (flux)

t_range = np.linspace(7200, t_used, 100) #issues here hm bc divide by 0
flux_val = [flux_calc(Dij_memb, t, c1, c0) for t in t_range]

plt.plot(t_range, flux_val)
plt.xlabel("Time (s)")
plt.ylabel("Flux (mol/cm^2*s)")
plt.title("21 mg Nicotine Patch: Flux vs. Time")
plt.grid(True)
plt.show()

# Uptake
def uptake_calc(conc, t):
  return (2 * area * conc * (Dij_memb * t_range / pi)**(1/2))

conc_val = np.linspace(c0/3, c0, 100)

conc_mesh, time_mesh = np.meshgrid(conc_val, t_range, indexing="ij")

uptake_val = uptake_calc(conc_mesh, time_mesh)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

surface = ax.plot_surface(conc_mesh, time_mesh, uptake_val, cmap = "plasma")

ax.set_xlabel("Concentration (mol/mL)")
ax.set_ylabel("Time (s)")
ax.set_zlabel("Nicotine Patch: Uptake")

fig.colorbar(surface, ax = ax, label = "Uptake")
plt.title("Uptake vs. Concentration vs. Time")
```
